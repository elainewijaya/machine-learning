# Part 1: Least Squares

1. Generate classification data of two classes.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Datapoints for both classes
N1 = 100
N2 = 100
# K is the number of classes
K = 2
sigma = 1.0

mean = (1,1)
# Covariance matrix with sigma as diagonal
cov = [[sigma, 0], [0, sigma]]
X1 = np.random.multivariate_normal(mean, cov, N1)
# create list of 'red' for the X1 points
c1 = ['red'] * len(X1)

mean = (5,5)
# Covariance matrix with sigma as diagonal
cov = [[sigma, 0], [0,sigma]]
X2 = np.random.multivariate_normal(mean,cov,N2)
c2 = ['blue']*len(X2)

# All X datapoints - first 50 are red, next 50 are blue
X = np.concatenate((X1,X2))
color = np.concatenate((c1,c2))


T = 0*np.ones([len(X), K])
for n in range(0, len(X)):
    if(n<len(X1)):
        T[n][0] = 1
    if(n>N1 and n<len(X1)+len(X2)):
        T[n][1] = 1
T = T.astype(int)

ones = np.array([np.ones(X.shape[0])])
X = np.concatenate((X, np.transpose(ones)), axis = 1)

plt.scatter(X[:,0], X[:,1], marker = 'o', c = color)
plt.show()
```

2. Write code to estimate matrix W using least-squares.
3. Draw the decision line, similar to:

```python
W = np.dot(np.dot(np.linalg.inv(np.dot(np.transpose(X), X)), np.transpose(X)), T)
Y = np.array([0,0])
for i in range(X.shape[0]):
    Y = np.vstack((Y,np.dot(np.transpose(W),X[i])))
Y = np.delete(Y,0,0)

plt.scatter(X[:,0], X[:,1], marker = 'o', c = color)
W1 = np.transpose(W)[0]
W2 = np.transpose(W)[1]

# Plot the resulting regression line for class 1
# Here, W1[2] is the w_0 constant
x1 = np.linspace(0, 9)
y1 = -W1[2] / W1[0] - (W1[1] / W1[0]) * x1

# Plot the resulting regression line for class 2
# Here, W2[2] is the w_0 constant
x2 = np.linspace(0, 9)
y2 =  -W2[2] / W2[0] - (W2[1] / W2[0]) * x2

plt.plot(x1,y1,linestyle='-', color = 'red')
plt.plot(x2,y2,linestyle='-', color = 'blue')

plt.show()

```

Increasing sigma makes the data more noisy over a larger area. This means that they overlap and, if one were to map out the "centre" of each data cluster, both data clusters would end up being closer to each other relative to the size of the clusters than before.
The decision lines are closer to one another overall and are still parallel. However, the decision lines are closer to the outer edges of each cluster of data. This is because while the lines want to get closer to one another due to the clusters getting closer to one another, they are still repelling away from each other because they are from separate classes. The compromise are lines that appear closer to the opposite class, but also further away at the edges of each dataset.

11. Explain what's wrong with the lines.

The blue cluster further away, the outlier, is heavily skewing the decision line. Since the least squares method works by minimizing the sum of the squared error, any training point that has a dependent value that differs a lot from the rest of the data will have a disproportionately large effect on the resulting constants that are being solved for. The blue outlier cluster is pulling the final decision line towards it, making the final decision line inacurrate since there are blue points on the red side of decision line.
